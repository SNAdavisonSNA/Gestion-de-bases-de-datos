Genero table 

```python
def extract_genres(x):
    if isinstance(x, list):
        # Lista de dicts o strings
        result = []
        for item in x:
            if isinstance(item, dict) and 'name' in item:
                result.append(item['name'])
            elif isinstance(item, str):
                result.append(item)
        return result
    elif isinstance(x, str):
        # Extraer palabras del string, útil si no hay comillas
        return re.findall(r'\b\w+\b', x)
    return []

df['genres_names'] = df['genres'].apply(extract_genres)

all_genres = [g for sublist in df['genres_names'] for g in sublist]
unique_genres = sorted(list(set(all_genres)))  # sorted para orden alfabético

df_genres = pd.DataFrame(unique_genres, columns=['genre'])
```

```python
df_genres.to_csv("df_genres.csv", index=False)
```

Platform

```python
def extract_genres(x):
    if isinstance(x, list):
        # Lista de dicts o strings
        result = []
        for item in x:
            if isinstance(item, dict) and 'name' in item:
                result.append(item['name'])
            elif isinstance(item, str):
                result.append(item)
        return result
    elif isinstance(x, str):
        # Extraer palabras del string, útil si no hay comillas
        return re.findall(r'\b\w+\b', x)
    return []

df['platform'] = df['platform'].apply(extract_genres)

all_platform = [g for sublist in df['platform'] for g in sublist]
unique_platform = sorted(list(set(all_platform)))  # sorted para orden alfabético

df_platform = pd.DataFrame(unique_platform, columns=['platform'])
```

Mapeo Game-Genre

```python
# Crear el mapeo juego-género usando la columna 'genres' del df principal
game_genre_mapping = []

for idx, row in df.iterrows():
    game_id = idx  # el id que agregaste como índice
    game_name = row['name']
    
    # Obtener la lista de géneros de la columna 'genres'
    genres_list = row['genres']
    
    # Convertir a lista si es string
    if isinstance(genres_list, str):
        import ast
        try:
            genres_list = ast.literal_eval(genres_list)
        except:
            # Limpiar el string si es necesario
            genres_list = genres_list.strip('[]').replace("'", "").split(', ')
            genres_list = [g.strip() for g in genres_list if g.strip()]
    
    # Para cada género en la lista del juego
    if isinstance(genres_list, list):
        for genre_name in genres_list:
            # Buscar el genre_id en df_genres
            genre_match = df_genres[df_genres['genre'] == genre_name]
            if not genre_match.empty:
                genre_id = genre_match.index[0]
                game_genre_mapping.append({
                    'game_id': game_id,
                    'game_name': game_name,
                    'genre_id': genre_id,
                    'genre_name': genre_name
                })

# Crear DataFrame con el mapeo
df_game_genre_mapping = pd.DataFrame(game_genre_mapping)

# Mostrar resultados
print("Mapeo juego-género (primeras 20 filas):")
print(df_game_genre_mapping.head(20))
print(f"\nTotal de relaciones: {len(df_game_genre_mapping)}")

# Guardar solo las columnas necesarias para SQL
df_game_genre_sql = df_game_genre_mapping[['game_id', 'genre_id']]
df_game_genre_sql.to_csv('game_genre_mapping.csv', index=False)

print("\nArchivo guardado: game_genre_mapping.csv")
print(df_game_genre_sql.head(10))
```

Mapeo Game-Platform 

```python
# Crear el mapeo juego-plataforma usando la columna 'platform' del df principal
game_platform_mapping = []

for idx, row in df.iterrows():
    game_id = idx  # el id que agregaste como índice
    game_name = row['name']
    
    # Obtener la lista de plataformas de la columna 'platform'
    platforms_list = row['platform']
    
    # Convertir a lista si es string
    if isinstance(platforms_list, str):
        import ast
        try:
            platforms_list = ast.literal_eval(platforms_list)
        except:
            # Limpiar el string si es necesario
            platforms_list = platforms_list.strip('[]').replace("'", "").split(', ')
            platforms_list = [p.strip() for p in platforms_list if p.strip()]
    
    # Para cada plataforma en la lista del juego
    if isinstance(platforms_list, list):
        for platform_name in platforms_list:
            # Buscar el platform_id en df_platform
            platform_match = df_platform[df_platform['platform'] == platform_name]
            if not platform_match.empty:
                platform_id = platform_match.index[0]
                game_platform_mapping.append({
                    'game_id': game_id,
                    'game_name': game_name,
                    'platform_id': platform_id,
                    'platform_name': platform_name
                })

# Crear DataFrame con el mapeo
df_game_platform_mapping = pd.DataFrame(game_platform_mapping)

# Mostrar resultados
print("Mapeo juego-plataforma (primeras 20 filas):")
print(df_game_platform_mapping.head(20))
print(f"\nTotal de relaciones: {len(df_game_platform_mapping)}")

# Guardar solo las columnas necesarias para SQL
df_game_platform_sql = df_game_platform_mapping[['game_id', 'platform_id']]
df_game_platform_sql.to_csv('game_platform_mapping.csv', index=False)

print("\nArchivo guardado: game_platform_mapping.csv")
print(df_game_platform_sql.head(10))
```

Conectar SQL

```python
import pandas as pd
from sqlalchemy import create_engine

# Conectar a MySQL
engine = create_engine("mysql+pymysql://root:12345@localhost:3306/videojuegos_db")

# Lista de archivos CSV y sus tablas
archivos = [
    ('df_sql.csv', 'games'),
    ('df_genres.csv', 'genres'),
    ('df_platform.csv', 'platforms'),
    ('game_genre_mapping.csv', 'game_genres'),
    ('game_platform_mapping.csv', 'game_platforms')
]

# Insertar cada archivo
for archivo, tabla in archivos:
    try:
        df = pd.read_csv(archivo)
        df.to_sql(tabla, engine, if_exists='append', index=False)
        print(f"✓ Insertados {len(df)} registros en {tabla}")
    except Exception as e:
        print(f"✗ Error en {tabla}: {e}")

print("Proceso completado")
```
