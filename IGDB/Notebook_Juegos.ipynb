```python
import pandas as pd
from sqlalchemy import create_engine

# Conectar a MySQL
engine = create_engine("mysql+pymysql://usuario:contraseña@localhost:3306/videojuegos_db")

# Lista de archivos CSV y sus tablas
archivos = [
    ('games.csv', 'games'),
    ('game_modes.csv', 'game_modes'),
    ('genres.csv', 'genres'),
    ('themes.csv', 'themes'),
    ('platforms.csv', 'platforms'),
    ('game_game_modes.csv', 'game_game_modes'),
    ('game_genres.csv', 'game_genres'),
    ('game_themes.csv', 'game_themes'),
    ('game_platforms.csv', 'game_platforms')
]

# Insertar cada archivo
for archivo, tabla in archivos:
    try:
        df = pd.read_csv(archivo)
        df.to_sql(tabla, engine, if_exists='append', index=False)
        print(f"✓ Insertados {len(df)} registros en {tabla}")
    except Exception as e:
        print(f"✗ Error en {tabla}: {e}")

print("Proceso completado")
```
